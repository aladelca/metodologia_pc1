```python
#Importar librerias
import pandas as pd
```

```python
#Cargar los datasets
general_data = pd.read_csv("../data/raw/general_data.csv")
survey_data = pd.read_csv("../data/raw/employee_survey_data.csv")
```

```python
#Mostrar las primeras 3 filas del DataFrame
general_data.head(3)
```

```python
#Mostrar las primeras 3 filas del DataFrame
survey_data.head(3)
```

```python
#Unir los DataFrames general_data y survey_data
merged_data = pd.merge(general_data, survey_data, on="EmployeeID", how="inner")
merged_data.head(3)
```

```python
#Calcular la columna average_employee_satisfaction 
survey_columns = survey_data.columns.drop("EmployeeID") 
merged_data["average_employee_satisfaction"] = merged_data[survey_columns].mean(axis=1)
```
