```python
# limpieza de contenido
ds_in_time_pivot = ds_in_time.melt(var_name = "fecha_ingreso", value_name = "registro_ingreso") # pivoteo del dataset in_time
ds_in_time_pivot = ds_in_time_pivot.dropna() # eliminacion de filas con valores nulos
ds_in_time_pivot["registro_ingreso"] = pd.to_datetime(ds_in_time_pivot["registro_ingreso"]) # conversion de la columna registro a tipo datetime
ds_in_time_pivot["fecha_ingreso"] = pd.to_datetime(ds_in_time_pivot["fecha_ingreso"]) # conversion de la columna fecha a tipo datetime
ds_in_time_pivot = ds_in_time_pivot.sort_values(["fecha_ingreso", "registro_ingreso"])
ds_in_time_pivot["id"] = range(1, len(ds_in_time_pivot) + 1)

# limpieza de contenido
ds_out_time_pivot = ds_out_time.melt(var_name="fecha_salida", value_name="registro_salida") # pivoteo del dataset in_time
ds_out_time_pivot = ds_out_time_pivot.dropna() # eliminacion de filas con valores nulos
ds_out_time_pivot["registro_salida"] = pd.to_datetime(ds_out_time_pivot["registro_salida"]) # conversion de la columna registro a tipo datetime
ds_out_time_pivot["fecha_salida"] = pd.to_datetime(ds_out_time_pivot["fecha_salida"]) # conversion de la columna fecha a tipo datetime
ds_out_time_pivot = ds_out_time_pivot.sort_values(["fecha_salida", "registro_salida"])
ds_out_time_pivot["id"] = range(1, len(ds_out_time_pivot) + 1)

ds_final = pd.merge(ds_in_time_pivot, ds_out_time_pivot, on=["id"], how="left") # merge de los dos datasets
```

```python
ds_final 
```

```python
fil = ds_final.shape[0] #cuantas filas tiene el dataset in_time 
col = ds_final.shape[1] #cuantas columnas tiene el dataset in_time 
print(f"Filas: {fil} y Columnas: {col}") #  1041930 y Columnas: 2
```

```python
ds_final['hora_ingreso_pactado'] = pd.to_datetime(ds_final["fecha_ingreso"].astype(str) + " 09:00:00")

ds_final["hora_registro_ingreso"] = ds_final['registro_ingreso'].dt.hour

dias = {
    "Monday": "Lunes"
    ,"Tuesday": "Martes"
    ,"Wednesday": "MiÃ©rcoles"
    ,"Thursday": "Jueves"
    ,"Friday": "Viernes"
    ,"Saturday": "SÃ¡bado"
    , "Sunday": "Domingo"
}
ds_final['diasemana_ingreso'] = ds_final['registro_ingreso'].dt.day_name().map(dias)
```

```python
ds_final["minutos_diferencia_ingreso"] = (ds_final["registro_ingreso"] - ds_final["hora_ingreso_pactado"]).dt.total_seconds() / 60
```

```python
def categorizar_tolerancia(minutos):
    if abs(minutos) <= 5:
        return "Puntual"
    elif 5 < minutos <= 15:
        return "Retraso leve"
    elif minutos > 15:
        return "Retraso grave"
    else:
        return "Llego antes del horario"

ds_final['categoria_tolerancia'] = ds_final['minutos_diferencia_ingreso'].apply(categorizar_tolerancia)
ds_final.tail(4)
```

```python
ds_final.groupby("categoria_tolerancia")["registro_ingreso"].count().reset_index()
```

# Generando las variables

```python
ds_final["horas_trabajadas"] = (ds_final["registro_salida"] - ds_final["registro_ingreso"]).dt.total_seconds() / 3600
ds_final.head(4)
```

```python
# Extraer el mes y aÃ±o
ds_final["mes"] = ds_final["registro_ingreso"].dt.month
ds_final["aÃ±o"] = ds_final["registro_ingreso"].dt.year

# Calcular promedio mensual de horas trabajadas
promedio_mensual = ds_final.groupby(["aÃ±o", "mes"])["horas_trabajadas"].transform("mean")
desviacion_estandar = ds_final.groupby(["aÃ±o", "mes"])["horas_trabajadas"].transform("std")

# Agregar la columna al DataFrame
ds_final["promedio_mensual_horas"] = promedio_mensual
ds_final["desviacion_estandar"] = desviacion_estandar


ds_final["tipo_jornada"] = ds_final["horas_trabajadas"].apply(lambda x: "Corta" if x < 4 else "Normal" if x < 8 else "Larga")
```

```python
def categorizar_horas_extra(horas):
    if horas <= 0:
        return "Sin horas extra"
    elif horas <= 1:
        return "Horas extra leves"
    elif horas <= 3:
        return "Horas extra moderadas"
    else:
        return "Horas extra altas"

# Aplicar la funciÃ³n al DataFrame
ds_final["categoria_horas_extra"] = ds_final["horas_extra"].apply(categorizar_horas_extra)
```
